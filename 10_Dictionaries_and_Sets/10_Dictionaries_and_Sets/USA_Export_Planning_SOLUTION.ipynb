### ‚Ä¢ Size Counts

üî∏ First I need **a dictionary** with the **item ids** in `items_dict` as **keys**, and **the number of sizes available for each item as values**.

üî∏ Sizes are stored as a Python list. 

```python
sizes_count = {}
```

```python
# Write your code here
for item_id, info in items_dict.items():
    sizes_count[item_id] = len(info[3])
    
print('Done!')
```

```python
sizes_count
```

```python
items_dict
```

### ‚Ä¢ New Size Options

```python
# Write your code here
new_items = {
    10010: 4,
    10011: 5
}

sizes_count.update(new_items)
```

### ‚Ä¢ U.S. Converted List

üî∏ Finally, pull the **price** out of `items_dict` and **create a list (`usd_prices`)** with converted U.S. pricing.

üî∏ Use an exchange rate of **1.08**.

```python
items_dict
```

```python
usd_prices = []
exchange_rate = 1.08
```

```python
# Write your code here
for value in items_dict.values():
    usd_price = round(exchange_rate * value[1], 2)
    usd_prices.append(usd_price)
    
print('Done!')
```
