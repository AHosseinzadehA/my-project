### • Importing NumPy, Pandas & Matplotlib

🔸 Import `numpy`, `pandas`, `matplotlib` here

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
````

````python
prices = [3.99, 4.23, 10.45, 20.66]
````

````python
# 1. Calculate the mean of 'prices'
np.mean(prices)
````

````python
items_dict  = {
    "customer_id": ['c00001', 'c00012', 'c00007', 'c00009'],
    "subtotal": [80, 5, 33, 70],
    "location": ["Germany", "Italy", "France", "Germany"]
}
````

````python
# 2. Create a Pandas DataFrame using 'items_dict'
pd.DataFrame(data=items_dict)
````

````python
x = [0, 1, 2, 3, 4, 5, 6]
y = [0, 1, 4, 9, 16, 25, 36]
````

````python
# 3. Create a line plot using 'x' & 'y'
plt.plot(x, y)
````

````python
items = ["Shirt", "Dress", "Hoodie"]
inventory = [21, 33, 8]
````

````python
# 4. Create a bar plot using 'items' & 'inventory'
plt.bar(items, inventory)
````

<a name="1-3"></a>

### A Professional Bar Plot 

🔸 We'll learn how to create this plot in [Matplotlib and Seaborn Zero to Hero Course](https://www.izlearn.ir/courses/matplotlib-seaborn-zero-to-hero-course/) 🔥

````python
plt.style.use("seaborn-v0_8")
fig, ax = plt.subplots(figsize=(6, 4), dpi=100)

ax.bar(items, inventory, color="#4776E6", width=0.5)

ax.set(title="Bar plot of items & items inventory",
       xlabel="Items",
       ylabel="Inventory")

ax.axhline(np.mean(inventory),
           linestyle='--',
           color="#8E54E9",
           lw=1)

# Save our fig to the current directory
plt.savefig("bar-plot-items-inventory.png")
plt.show()
````
